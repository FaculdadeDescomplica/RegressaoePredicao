```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# variável preditora
X = np.array([ 220, 220, 220, 220, 220, 225, 225, 225, 225, 225, 230, 230, 230, 230, 230, 235, 235, 235, 235, 235 ])
# variável alvo
y = np.array([ 137, 137, 137, 136, 135, 135, 133, 132, 133, 133, 128, 124, 126, 129, 126, 122, 122, 122, 119, 122 ])
```

```python
# é necessário adicionar uma constante a matriz X
X_sm = sm.add_constant(X)
```

```python
# OLS vem de Ordinary Least Squares e o método fit irá treinar o modelo
results = sm.OLS(y, X_sm).fit()
```

```python
# mostrando as estatísticas do modelo
print(results.summary())
```

```python
clientes = pd.read_csv("/content/EcommerceCustomers.csv")
```

```python
sns.pairplot(clientes)
```

```python
sns.lmplot(x='Length of Membership', y='Yearly Amount Spent', data=clientes)
```

```python
y = clientes['Yearly Amount Spent']
```

```python
X= clientes[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test =train_test_split(X, y, test_size=0.3,random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test)
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
